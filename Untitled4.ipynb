```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.figure(figsize=(6,3),dpi=200)
x=[1,2,3,4,5,6,7,8]
y=[2,4,6,8,10,12,14,16]
#plt.plot(x,y ,label='y=2x',color='red',linewidth=3,marker='.',markeredgecolor='purple',markersize=10)
# shorthand[color][marker][line]
plt.plot(x,y, 'r^--',label='y=2x')

#lineno2
x2=np.arange(0,10,0.5)
x2
plt.plot( x2[:6],x2[:6]*x2[:6],'b',label='y=x^2')
plt.plot( x2[5:],x2[5:]*x2[5:],'b--',label='y=x^2')


plt.title("our first graph", fontdict={'fontsize':20})

plt.xlabel("x axis",fontdict={'fontsize':10})
plt.ylabel("y axis",fontdict={'fontsize':10})
plt.xticks([0,2,3,4,5,])
#plt.yticks([0,2,4,6,8,10,12])
plt.legend(loc="lower right", title="my lagend",fontsize="10")
plt.savefig('mygraph.png',dpi=200)
 

plt.show()



```
