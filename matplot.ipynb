```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
gas = pd.read_csv("C:\\Users\\ishig\\anaconda_projects\\gas_prices.csv")
```

```python

plt.figure(figsize=(8,5))
plt.title("Gas Prices over world in dollars")
#plt.plot(gas.Year,gas.Canada ,'b.-',label='canada')
#plt.plot(gas.Year,gas.USA,'r>--',label='United states')
#plt.plot(gas.Year,gas['South Korea'],'go-',label='south korea')
#plt.plot(gas.Year,gas.Australia,'y.--',label='Australia')
for country in gas.columns[1:]:
  
  plt.plot( gas.Year,gas[country],marker='.',label=country)
plt.legend()
plt.xticks(gas.Year[::6].tolist()+[2011])
plt.xlabel("year")
plt.ylabel("us dollars")
plt.savefig("gasprice.png")

plt.show()
```
