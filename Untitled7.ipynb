```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import LeaveOneOut
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler



x = iris.data  
y = iris.target 

df = pd.DataFrame(x, columns=iris.feature_names)
df['Species'] = y

print("Dataset Preview:")
print(df.head())


scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

print("Shape of x_scaled:", x_scaled.shape)


loo = LeaveOneOut()


y_true, y_pred = [], []
y_prob = []


classifier_SVM = SVC(kernel='rbf', random_state=0, probability=True)


sumAcc = 0
n_splits = 0  


for fold, (train_index, test_index) in enumerate(loo.split(x_scaled), start=1):
    print(f"Fold {fold}:")
   
   
    x_train, x_test = x_scaled[train_index], x_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    classifier_SVM.fit(x_train, y_train)
    
   
    yhat = classifier_SVM.predict(x_test)
    yprob = classifier_SVM.predict_proba(x_test)

    
    y_true.append(y_test[0]) 
    y_pred.append(yhat[0])
    y_prob.append(yprob[0])

  
    acc = accuracy_score(y_true, y_pred)
    print(f"Accuracy for Fold {fold}: {acc:.3f}")

    sumAcc += acc
    n_splits += 1 

overall_accuracy = sumAcc / n_splits
print(f'Overall Accuracy: {overall_accuracy:.3f}')


svm_prc_loo = precision_score(y_true, y_pred, average='weighted', zero_division=0)
svm_f1_loo = f1_score(y_true, y_pred, average='weighted')

try:
    
    svm_auc_loo = roc_auc_score(y_true, np.array(y_prob), multi_class='ovr')
except ValueError:
    svm_auc_loo = None 

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix is:")
print(cm)


if len(np.unique(y_true)) > 1:  
    plt.figure(figsize=(10, 8))

    
    for i in range(3):
        fpr, tpr, _ = roc_curve(y_true, np.array(y_prob)[:, i], pos_label=i)
        plt.plot(fpr, tpr, linestyle='--', label=f'Class {i}')

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray') 
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve (One-vs-Rest) for Each Class')
    plt.legend(loc='best')
    plt.show()

    print(f"ROC AUC Score: {svm_auc_loo:.3f}")
else:
    print("Skipping ROC AUC calculation due to a single class in the fold.")
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, precision_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target  

loo = LeaveOneOut()


y_true, y_pred = list(), list()
y_prob = np.empty((0, 3)) 


classifier_DTC = DecisionTreeClassifier(max_depth=3)

sumAcc = 0


for train_index, test_index in loo.split(x):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]

    
    classifier_DTC.fit(x_train, y_train)

   
    yhat = classifier_DTC.predict(x_test)
    yprob = classifier_DTC.predict_proba(x_test) 

    y_true.append(y_test[0])
    y_pred.append(yhat[0])
    y_prob = np.vstack([y_prob, yprob[0]]) 


    acc = accuracy_score(y_true, y_pred)
    sumAcc += acc


import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelBinarizer


X = np.random.rand(100, 5) 
y = np.random.randint(0, 3, 100)  


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)


y_prob = clf.predict_proba(X_test)

# Binarize the true labels (necessary for multi-class ROC AUC calculation)
lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)

# Compute ROC curve and ROC AUC for each class
fpr, tpr, thresholds = {}, {}, {}
roc_auc = {}

# One-vs-Rest (OvR) ROC curve for each class
for i in range(y_prob.shape[1]):  # For each class
    fpr[i], tpr[i], thresholds[i] = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])


plt.figure(figsize=(8, 6))


for i in range(y_prob.shape[1]):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')


plt.plot([0, 1], [0, 1], 'k--', label='Random classifier')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-Class ROC Curve')
plt.legend(loc='best')

plt.show()

for i in range(y_prob.shape[1]):
    print(f'Class {i} AUC: {roc_auc[i]:.3f}')

overall_accuracy = sumAcc / len(y)
print(f'Overall Accuracy: {overall_accuracy:.3f}')

dtc_f1_loo = f1_score(y_true, y_pred, average='weighted')
dtc_auc_loo = roc_auc_score(y_true, y_prob, multi_class='ovr')  # Multi-class AUC


cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)

plt.figure(figsize=(10, 8))

for i in range(3):  
    fpr, tpr, _ = roc_curve(y_true, y_prob[:, i], pos_label=i)
    plt.plot(fpr, tpr, linestyle='--', label=f'Class {i}')


plt.plot([0, 1], [0, 1], linestyle='--', color='gray') 
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (One-vs-Rest) for Each Class')
plt.legend(loc='best')
plt.show()


print(f"ROC AUC Score: {dtc_auc_loo:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelBinarizer


X = np.random.rand(100, 5)  
y = np.random.randint(0, 3, 100) 

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)


y_prob = clf.predict_proba(X_test)


lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)


fpr, tpr, thresholds = {}, {}, {}
roc_auc = {}


for i in range(y_prob.shape[1]): 
    fpr[i], tpr[i], thresholds[i] = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(8, 6))


for i in range(y_prob.shape[1]):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random classifier')


plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-Class ROC Curve')
plt.legend(loc='best')

plt.show()


for i in range(y_prob.shape[1]):
    print(f'Class {i} AUC: {roc_auc[i]:.3f}')
```
