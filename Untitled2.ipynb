```python
import urllib.request
import numpy as np



# Download the file
urllib.request.urlretrieve('https://raw.githubusercontent.com/KcPele/data_analysis_with_python_jovian/master/climate.txt', 'climate.txt')

# Load the data (adjust delimiter based on file format)
climate_data = np.genfromtxt('climate.txt',delimiter=',',skip_header=1)  # If CSV

climate_data
```

```python
w1,w2,w3=0.3,0.2,0.5
weights=np.array([w1,w2,w3])
yields =climate_data @ weights
yields
```

```python
yields.shape
```

```python
results=np.concatenate((climate_data, yields.reshape(10000,1)),axis=1)
results
```
