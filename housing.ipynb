```python

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

val=pd.read_csv("C:\\Users\\ishig\\anaconda_projects\\Housing.csv")
val
val.describe()
val=pd.read_csv("C:\\Users\\ishig\\anaconda_projects\\Housing.csv")
val

from sklearn.linear_model import LinearRegression
model=LinearRegression()
features = ['mainroad', 'guestroom', 'basement', 'airconditioning']
import numpy as np
for feature in features:
    val[feature] = np.where(val[feature] == 'yes', 1, 0)
x = val[['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'mainroad', 'guestroom', 'basement', 'airconditioning']]
y = val['price'] 
model.fit(x, y)
   # model.predict(np .array([[7800,2,2,1,1,0,1,0,1]])
    
# sample_input = pd.DataFrame([[7800, 2, 2, 1, 1, 0, 1, 1, 1]])
val['predicted_price'] = model.predict(x)


print(val.head(10))

plt.figure(figsize=(8, 6))
sns.lineplot(x=val['price'], y=val['predicted_price'])
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Housing Prices")
plt.show()

from sklearn.metrics import mean_squared_error
loss=np.sqrt(mean_squared_error(val['price'],val['predicted_price']))
loss
    
```

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
val=pd.read_csv("C:\\Users\\ishig\\anaconda_projects\\Housing.csv")
sns.barplot(data=val,x='airconditioning',y='price');
```

```python


fur_map = {'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0}
val['furnishingstatus'] = val['furnishingstatus'].map(fur_map)
val
```
